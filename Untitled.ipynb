```python
# it picks a specific col as a index
pd.read_csv('aug_train.csv', index_col='enrollee_id')
```

```python
#header is used when the we want to mention the header (a specific row) so that the pandas makes no mistake in picking the header 
# by default the pandas will read the first row as header if the data set taken has the header at the 2nd row then pandas will read the file incorrectly
#and that's why we give the header parameter
pd.read_csv('aug_train.csv',header=1)
```

```python
#usecols will take only the col mentioned in the array and it will read only that col's
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

```python
#dtype is used to convert type of the coloum
pd.read_csv('aug_train.csv', dtype={'target':int}).info()
```
